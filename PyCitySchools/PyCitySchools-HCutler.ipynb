```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools

total_schools = len(school_data_complete["school_name"].unique())
```

```python
#Calculate the total number of students

total_students = (school_data_complete["student_name"].count())
```

```python
#Calculate the total budget

total_budget = school_data["budget"].sum()
```

```python
#Calculate the average math score

average_math_score = school_data_complete["math_score"].mean()
```

```python
#Calculate the average reading score

average_reading_score = school_data_complete["reading_score"].mean()
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

students_passing_math_num = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()
students_passing_math_perc = (students_passing_math_num / total_students)* 100
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)

students_passing_reading_num = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["reading_score"].count()
students_passing_reading_perc = (students_passing_reading_num / total_students)* 100
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)

students_passing_m_and_r_num = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]["size"].count()
students_passing_m_and_r_perc = (students_passing_m_and_r_num / total_students)* 100
```

```python
#Create a dataframe to hold the above results

district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Percentage Passing Math": [students_passing_math_perc],
    "Percentage Passing Reading": [students_passing_reading_perc],
    "Percentage Passing Overall": [students_passing_m_and_r_perc]
})
district_summary_df
```

```python
#% passing reading (the percentage of students who passed reading)

students_passing_reading_num_2 = school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()
students_passing_reading_perc_2 = (students_passing_reading_num_2 / total_students_2) * 100
```

```python
#Highest performing schools

highest_performing_df = school_summary_df.sort_values("Percentage Passing Overall", ascending = False)
highest_performing_df.head(5)
```

```python
#Scores by school spending

spending_bins = [0, 584, 629, 644, 675]
spending_bin_labels = ["<$585", "$585-630", "$630-645", "$645-650"]

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=spending_bin_labels)
scores_by_spending = school_summary_df.groupby("Spending Ranges (Per Student)")

average_math_score_by_spending = school_summary_df.groupby("Spending Ranges (Per Student)")["Average Math Score"].mean()
average_reading_score_by_spending = school_summary_df.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()
students_passing_math_by_spend_perc = school_summary_df.groupby("Spending Ranges (Per Student)")['Percentage Passing Math'].mean()
students_passing_reading_by_spend_perc = school_summary_df.groupby("Spending Ranges (Per Student)")['Percentage Passing Reading'].mean()
students_passing_m_and_r_by_spend_perc = school_summary_df.groupby("Spending Ranges (Per Student)")['Percentage Passing Overall'].mean()
```

```python
#Scores by school spending df

scores_by_spending_df = pd.DataFrame({
    "Average Math Score": average_math_score_by_spending,
    "Average Reading Score": average_reading_score_by_spending,
    "Percentage Passing Math": students_passing_math_by_spend_perc,
    "Percentage Passing Reading": students_passing_reading_by_spend_perc,
    "Percentage Passing Overall": students_passing_m_and_r_by_spend_perc
      })
scores_by_spending_df
```
